**Ejemplo de Interacción con API**

**`Importar Librerías`**

```python
import requests
import pandas as pd
import matplotlib.pyplot as plt
```

**URL de la API Pública de (OpenWeatherMap para calidad de aire  de una ciudad específica como, Madrid)**. Normalmente, se necesitaría un API Key, pero por simplicidad usamos un endpoint abierto

```python
url = "https://api.open-meteo.com/v1/forecast?latitude=40.4168&longitude=-3.7038&hourly=temperature_2m"
```

**Enviar la petición a la API**



```python
response = requests.get(url)
```

**Check if the response was successful**




```python
if response.status_code == 200:
    data = response.json()  # Parse the response JSON
    hourly_data = data['hourly']  # Extract the hourly forecast data
    df = pd.DataFrame(hourly_data)  # Convert to a DataFrame for better handling

    # Convert the time to a pandas datetime object for better plotting
    df['time'] = pd.to_datetime(df['time'])

    # Visualize the data (temperature over time)
    plt.figure(figsize=(10, 6))
    plt.plot(df['time'], df['temperature_2m'], marker='o')
    plt.title("Temperature Forecast in Madrid (Hourly)", fontsize=14)
    plt.xlabel("Time", fontsize=12)
    plt.ylabel("Temperature (°C)", fontsize=12)
    plt.grid(True)
    plt.xticks(rotation=45)
    plt.tight_layout()

    # Display the plot
    plt.show()
else:
    print("Failed to retrieve data from the API. Status code:", response.status_code)
```
